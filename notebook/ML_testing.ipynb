```python
RandomForestClassifier for ASL

RandomForest uses Decision trees as the foundation of how it works
A decision tree is a model that makes predictions by asking a sequence of simple questions about the input features.
    The tree keeps splitting the data into smaller groups based on the feature
Random Forest instead of using one giant decision tree, it uses many smaller decision trees that work on random subsets of the training data
 For classification:
   - Each tree votes for a class.
   - The forest prediction is the **majority vote** across all trees.
Useful as the trees are able to make small mistakes and are averaged out by the end to also reduce the chance of overfitting.
       Within our data, this is useful by being able to help with possible non-linear patterns, which is helpful when relationships between features and labels are not simple lines.

```
