```python
## 3. Null baselines

Before training real models, it's good to see how a **dummy model** performs.

We use:

- Majority baseline (always predicts the most common class).
- Stratified baseline (predicts classes at random, but following their frequency).

If our real models **do not beat** these baselines, then something is wrong.
```
